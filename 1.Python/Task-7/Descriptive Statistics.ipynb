<h1> Mean </h1>
<hr>
<h3>Definition:</h3>

<p>Mean is the <i>average of the given numbers </i> and is calculated by dividing the sum of given numbers by the total number of numbers. </p>


$$
\overline X = \frac{\Sigma x}{n}\\
$$
<br>
<p>
where, <br>
&emsp;&emsp;&emsp;&emsp;$\overline X $= Mean <br>
&emsp;&emsp;&emsp;&emsp;$\Sigma x$ = Sum of all variables <br>
&emsp;&emsp;&emsp;&emsp;n = Total number of variables <br>
</p>
<br>
<h3>Example:</h3>

What is the mean of 2, 4, 6, 8 and 10?

<u>Solution:</u>

First, add all the numbers.

2 + 4 + 6 + 8 + 10 = 30

Now divide by 5 (total number of observations).

Mean = 30/5 = 6

```python
data.Mthly_HH_Income.mean()
```

<a id='median'></a>
<h1> Median </h1>
<hr>
<h3>Definition:</h3>

<p>Median is defined as the <i>middle value of a sorted list </i> of numbers</p>

<p>If the given number of observations/data is odd, then the formula to calculate the median is:</p>

$$
Median = \{{(n+1)}/{2}\}^{th} term\\
$$
<p>If the given number of observations is even, then the formula to find the median is given by:</p>

\$$
Median = [(n/2)^{th} term + \{{(n+1)}/{2}\}^{th} term]/2
$$

<br>
<p>
where, <br>
&emsp;&emsp;&emsp;&emsp;n = Total number of variables <br>
</p>
<br>
<h3>Example 1:</h3>

Determine the median for the given dataset:

5, 7, 4, 8, 6

<u>Solution</u>:

Given dataset: 5, 7, 4, 8, 6

Here, the number of observations is odd, i.e., 5 observations are given.

n = 5

Now, arrange the numbers in ascending order

4, 5, 6, 7, 8

The formula to calculate the median for odd observations is:

Median = {(n+1)/2}th term

Median = {(5+1)/2}th term

Median = 3rd term

Here, the 3rd term is 6.

Therefore, the median for the given dataset is 6.
<br>
<br>

<h3>Example 2:</h3>

Determine the median for the given dataset:

4, 7, 3, 8, 6, 2

<u>Solution :</u>

Given dataset: 4, 7, 3, 8, 6, 2

Here, the number of observations is even, i.e., 6 observations are given.

n = 6

Now, arrange the numbers in ascending order

2, 3, 4, 6, 7, 8

The formula to calculate the median for odd observations is:

Median = [(n/2)th term + {(n/2)+1}th term]/2

Median = [(6/2)th term + {(6/2)+1}th term]/2

Median = (3rd term + 4th term)/2

Here, the 3rd term is 4 and the 4th term is 6

Therefore, median = (4+6)/2

= 10/2 = 5

Therefore, the median for the given dataset is 5.

```python
data['Mthly_HH_Expense'].median()
```

```python
plt.figure(figsize = (7,5))
x_med = np.median(data.Mthly_HH_Expense)
k = sns.kdeplot(data.Mthly_HH_Expense, shade=True)
plt.axvline(x_med, c='red')

plt.show()
```

<h1> Variance </h1>
<hr>
<h3>Definition:</h3>

<p>The variance is a measure of how far a set of data are  <i>dispersed out from their mean or average value</i></p>
<br>
<img src = https://blog.minitab.com/hubfs/Imported_Blog_Media/within_group_dplot.png>
<h3>Properties of Variance</h3>

<p>It is always non-negative since each term in the variance sum is squared and therefore the result is either positive or zero.</p>
<p>Variance always has squared units. For example, the variance of a set of weights estimated in kilograms will be given in kg squared. Since the population variance is squared, we cannot compare it directly with the mean or the data themselves.</p>

<h5>Population Formula</h5>
<p>The population variance formula is given by:</p>
$$
\sigma^{2} = \frac{1}{N} \sum_{i=1}^{N}(X_i-\mu)^{2}\\
$$
<br>
<p>
where, <br>
&emsp;&emsp;&emsp;&emsp;$\sigma^{2} $= Population variance <br>
&emsp;&emsp;&emsp;&emsp;N = Number of observations in population <br>
&emsp;&emsp;&emsp;&emsp;$X_i$ = $i^{th}$ observation in the population <br>
&emsp;&emsp;&emsp;&emsp;$\mu$ = Population mean <br>
</p>
<br>
    
<h5>Sample Formula</h5>
<p>The sample variance formula is given by:</p>
$$
s^{2} = \frac{1}{n-1} \sum_{i=1}^{N}(x_i-\overline x)^{2}\\
$$
<br>
<p>
where, <br>
&emsp;&emsp;&emsp;&emsp;$s^2$ = Sample variance <br>
&emsp;&emsp;&emsp;&emsp;n = Number of observations in sample <br>
&emsp;&emsp;&emsp;&emsp;$x_i$ = $i^{th}$ observation in the sample <br>
&emsp;&emsp;&emsp;&emsp;$\overline x$ = Sample mean <br>
</p>
<br>
    
<h3>Example 1:</h3>

If a die is rolled, then find the variance and standard deviation of the possibilities.

<u>Solution:</u> 

When a die is rolled, the possible outcome will be 6. So the sample space, n = 6 and the data set = { 1;2;3;4;5;6}.

To find the variance, first, we need to calculate the mean of the data set.

Mean, x = (1+2+3+4+5+6)/6 = 3.5

We can put the value of data and mean in the formula to get;

2 =  (xi  x)2/n

2 =  (6.25+2.25+0.25+0.25+2.25+6.25)

2 = 2.917

```python
print(data.var()['Annual_HH_Income'])
```

<hr>
<h3>Definition:</h3>

<p>The <i>spread of statistical data</i> is measured by the standard deviation. Distribution measures the deviation of data from its mean or average position.The degree of dispersion is computed by the method of estimating the deviation of data points. It is denoted by the symbol, .</p>

<img src =https://www.scribbr.de/wp-content/uploads/2020/09/standard-deviation-in-normal-distributions.png width = 700px height = 700px>
    
<h3>Properties of Standard Deviation</h3>

<p>It describes the square root of the mean of the squares of all values in a data set and is also called the root-mean-square deviation.</p>
<p>The smallest value of the standard deviation is 0 since it cannot be negative.</p>
<p>When the data values of a group are similar, then the standard deviation will be very low or close to zero. But when the data values vary with each other, then the standard variation is high or far from zero.</p>

<h5>Population Formula</h5>
<p>The population standard deviation formula is given as:</p>
$$
\sigma = \sqrt{\frac{1}{N} \sum_{i=1}^{N}(X_i-\mu)^{2}}\\
$$
<br>
<p>
where, <br>
&emsp;&emsp;&emsp;&emsp;$\sigma $= Population variance <br>
&emsp;&emsp;&emsp;&emsp;N = Number of observations in population <br>
&emsp;&emsp;&emsp;&emsp;$X_i$ = $i^{th}$ observation in the population <br>
&emsp;&emsp;&emsp;&emsp;$\mu$ = Population mean <br>
</p>
<br>
    
<h5>Sample Formula</h5>
<p>The sample standard deviation formula is:</p>
$$
s = \sqrt{\frac{1}{n-1} \sum_{i=1}^{N}(x_i-\overline x)^{2}}\\
$$
<br>
<p>
where, <br>
&emsp;&emsp;&emsp;&emsp;$s$ = Sample variance <br>
&emsp;&emsp;&emsp;&emsp;n = Number of observations in sample <br>
&emsp;&emsp;&emsp;&emsp;$x_i$ = $i^{th}$ observation in the sample <br>
&emsp;&emsp;&emsp;&emsp;$\overline x$ = Sample mean <br>
</p>
<br>

To find the variance, first, we need to calculate the mean of the data set.

<h1> Correlation </h1>
<hr>
<h3>Definition:</h3>

<p>Correlation refers to a process for establishing the <i>relationships between two variables.</i></p>
<p>In statistics, Correlation studies and measures the direction and extent of relationship among variables, so the correlation measures co-variation, not causation. Therefore, we should never interpret correlation as implying cause and effect relation. For example, there exists a correlation between two variables X and Y, which means the value of one variable is found to change in one direction, the value of the other variable is found to change either in the same direction (i.e. positive change) or in the opposite direction (i.e. negative change). Furthermore, if the correlation exists, it is linear, i.e. we can represent the relative movement of the two variables by drawing a straight line on graph paper.</p>

<h3>Types of Correlation</h3>

<p>The scatter plot explains the correlation between the two attributes or variables. It represents how closely the two variables are connected. There can be three such situations to see the relation between the two variables </p>

- Positive Correlation  when the values of the two variables move in the same direction so that an increase/decrease in the value of one variable is followed by an increase/decrease in the value of the other variable.
- Negative Correlation  when the values of the two variables move in the opposite direction so that an increase/decrease in the value of one variable is followed by decrease/increase in the value of the other variable.
- No Correlation  when there is no linear dependence or no relation between the two variables.

<br>
<img src=https://cqeacademy.com/wp-content/uploads/2018/06/Scatter-Plots-and-Correlation-Examples.png width = 700px height = 700px>
<br>
    
<h3>Correlation Coefficient</h3>

<p>The correlation coefficient, r, is a summary measure that describes the extent of the statistical relationship between two interval or ratio level variables. The correlation coefficient is scaled so that it is always between -1 and +1. When r is close to 0 this means that there is little relationship between the variables and the farther away from 0 r is, in either the positive or negative direction, the greater the relationship between the two variables.</p><br>

$$
r_{xy} = \frac{n\sum_{i=1}^{n}x_iy_i - \sum_{i=1}^{n}x_i \sum_{i=0}^{n}y_i} {\sqrt{n \sum_{i=1}^{n}x_i^2 - (\sum_{i=1}^{n}x_i)^2} \sqrt{n\sum_{i=1}^{n}y_i^2 - (\sum_{i=1}^{n}y_i)^2}}\\
$$
<br>
<p>
where, <br>
&emsp;&emsp;&emsp;&emsp;n = Quantity of Information<br>
&emsp;&emsp;&emsp;&emsp;$\sum x_i$ = Total of the First Variable Value <br>
&emsp;&emsp;&emsp;&emsp;$\sum y_i$ = Total of the Second Variable Value <br>
&emsp;&emsp;&emsp;&emsp;$\sum x_iy_i$ = Sum of the Product of first and Second Value<br>
&emsp;&emsp;&emsp;&emsp;$\sum x^2_i$ = Sum of the Squares of the First Value <br>
&emsp;&emsp;&emsp;&emsp;$\sum y^2_i$ = Sum of the Squares of the Second Value<br>
</p>
<br>

<h5>Population Formula</h5>
<p>The population correlation coefficient is given as:</p><br>
$$
r_{xy}=S_{xy}/S_xS_y
$$
<br>
<p>
where, <br>
&emsp;&emsp;&emsp;&emsp;$S_x$= sample standard deviation <br>
&emsp;&emsp;&emsp;&emsp;$S_y$ = sample standard deviation <br>
&emsp;&emsp;&emsp;&emsp;$S_{xy}$ = sample covariance <br>
</p>
<br>
    
<h5>Sample Formula</h5>
<p>The sample correlation coefficient is:</p><br>
$$
r_{xy}=\sigma_{xy}/\sigma_x\sigma_y
$$
<br>
<p>
where, <br>
&emsp;&emsp;&emsp;&emsp;$\sigma_x$= population standard deviation <br>
&emsp;&emsp;&emsp;&emsp;$\sigma_y$ = population standard deviation <br>
&emsp;&emsp;&emsp;&emsp;$\sigma_{xy}$ = population covariance <br>
</p>
<br>

<h1> Normal Distribution </h1>
<hr>
<h3>Definition:</h3>

<p>In probability theory and statistics, the Normal Distribution, also called the <i>Gaussian Distribution</i>, is the most significant continuous probability distribution. Sometimes it is also called a <i>bell curve</i>. A large number of random variables are either nearly or exactly represented by the normal distribution, in every physical science and economics. Furthermore, it can be used to approximate other probability distributions.</p>

<br>
<img src =https://miro.medium.com/max/1200/1*IdGgdrY_n_9_YfkaCh-dag.png width = 700px height = 700px> 
<br> 

<p>The Normal Distribution is defined by the probability density function for a continuous random variable in a system. Let us say, f(x) is the probability density function and X is the random variable. Hence, it defines a function which is integrated between the range or interval (x to x + dx), giving the probability of random variable X, by considering the values between x and x+dx.

f(x)  0  x  (,+)

And -+ f(x) = 1</p><br>

<p>The probability density function of normal or gaussian distribution is given by;</p>

$$
f(x,\mu,\sigma) = \frac{1}{\sigma\sqrt{2\pi}}e^{\frac{-(x-\mu)^2}{2\sigma^2}}
$$
<br>
<p>
where, <br>
&emsp;&emsp;&emsp;&emsp;x = variable <br>
&emsp;&emsp;&emsp;&emsp;$\mu$ = mean <br>
&emsp;&emsp;&emsp;&emsp;$\sigma$ = standard deviation <br>
</p>
<br>

<h3>Normal Distribution Standard Deviation</h3>
<p>Generally, the normal distribution has any positive standard deviation. We know that the mean helps to determine the line of symmetry of a graph, whereas the standard deviation helps to know how far the data are spread out. If the standard deviation is smaller, the data are somewhat close to each other and the graph becomes narrower. If the standard deviation is larger, the data are dispersed more, and the graph becomes wider. The standard deviations are used to subdivide the area under the normal curve. Each subdivided section defines the percentage of data, which falls into the specific region of a graph.</p>

<p>Using 1 standard deviation, the Empirical Rule states that,</p>

- Approximately 68% of the data falls within one standard deviation of the mean. (i.e., Between Mean- one Standard Deviation and Mean + one standard deviation)
- Approximately 95% of the data falls within two standard deviations of the mean. (i.e., Between Mean- two Standard Deviation and Mean + two standard deviations)
- Approximately 99.7% of the data fall within three standard deviations of the mean. (i.e., Between Mean- three Standard Deviation and Mean + three standard deviations)

<h1>Feature of Normal Distribution</h1>
<hr>
<p>Some of the important properties of the normal distribution are listed below:</p>

- In a normal distribution, the mean, median and mode are equal.(i.e., Mean = Median= Mode).
- The total area under the curve should be equal to 1.
- The normally distributed curve should be symmetric at the centre.
- There should be exactly half of the values are to the right of the centre and exactly half of the values are to the left of the centre.
- The normal distribution should be defined by the mean and standard deviation.
- The normal distribution curve must have only one peak. (i.e., Unimodal)
- The curve approaches the x-axis, but it never touches, and it extends farther away from the mean.

<h1> Positively Skewed & Negatively Skewed Normal Distribution </h1>
<hr>
<h3>Definition:</h3>

<p>Skewness is a measure of the asymmetry of a distribution. A distribution is asymmetrical when its left and right side are not mirror images.

A distribution can have right (or positive), left (or negative), or zero skewness. A right-skewed distribution is longer on the right side of its peak, and a left-skewed distribution is longer on the left side of its peak:</p>
<br><img src = https://cdn.scribbr.com/wp-content/uploads/2022/05/Skewness-of-a-distribution-Large.webp width = 900px height = 900px><br>

<h3>Zero Skew</h3>

<p>When a distribution has zero skew, it is symmetrical. Its left and right sides are mirror images.

Normal distributions have zero skew, but theyre not the only distributions with zero skew. Any symmetrical distribution, such as a uniform distribution or some bimodal (two-peak) distributions, will also have zero skew.</p>
<p>The distribution is approximately symmetrical, with the observations distributed similarly on the left and right sides of its peak. Therefore, the distribution has approximately zero skew.In a distribution with zero skew, the mean and median are equal.</p><br>

$$Zero-skew: Mode = Median = Mean$$

<br>
<br>
For example, the weights of six-week-old chicks are shown in the histogram below.
<table><tr><td><img src='https://cdn.scribbr.com/wp-content/uploads/2022/05/Zero-skew-symmetrical.webp'></td><td><img src='https://i.dlpng.com/static/png/4783342-bell-curve-overview-characteristics-and-uses-in-finance-bell-curve-png-1024_513_preview.png' width = 1500px height=1500px></td></tr></table>
<br>
    
<h3>Positive(Right) Skew</h3>

<p>A right-skewed distribution is longer on the right side of its peak than on its left. Right skew is also referred to as positive skew.

You can think of skewness in terms of tails. A tail is a long, tapering end of a distribution. It indicates that there are observations at one of the extreme ends of the distribution, but that theyre relatively infrequent. A right-skewed distribution has a long tail on its right side.</p>

<p>The mean of a right-skewed distribution is almost always greater than its median. Thats because extreme values (the values in the tail) affect the mean more than the median.</p><br>

$$Right-skew: Mean  >  Median  >  Mode$$

<br>
<br>
For example,the number of sunspots observed per year, shown in the histogram below, is an example of a right-skewed distribution. The sunspots, which are dark, cooler areas on the surface of the sun, were observed by astronomers between 1749 and 1983.
<table><tr><td><img src='https://cdn.scribbr.com/wp-content/uploads/2022/05/Right-skew-Positive-skew.webp'></td><td><img src='https://cdn.corporatefinanceinstitute.com/assets/Positively-Skewed-Distribution.png' width = 1500px height=1500px></td></tr></table>
<br>

<h3>Negative(Left) Skew</h3>

<p>A left-skewed distribution is longer on the left side of its peak than on its right. In other words, a left-skewed distribution has a long tail on its left side. Left skew is also referred to as negative skew.</p>

<p>The mean of a left-skewed distribution is almost always less than its median.</p><br>

$$Left-skew: Mode  >  Median  > Mean$$

<br>
<br>
For example,test scores often follow a left-skewed distribution, with most students performing relatively well and a few students performing far below average. The histogram below shows scores for the zoology portion of a standardized test taken by Indian students at the end of high school.
<table><tr><td><img src='https://cdn.scribbr.com/wp-content/uploads/2022/05/Left-skew-Negative-skew.webp'></td><td><img src='https://cdn.corporatefinanceinstitute.com/assets/negatively-skewed-distribution.png'></td></tr></table>
<br>
<br>

<p>The skewness is given by;</p>
$$
\tilde \mu_3 = \frac{\sum_i^N(X_i-\overline X)^3}{(N-1)*\sigma^3}\\
$$
<br>
<p>
where, <br>
&emsp;&emsp;&emsp;&emsp;$\tilde \mu_3$ = Skewness <br>
&emsp;&emsp;&emsp;&emsp;N = Number of Variables in the distribution<br>
&emsp;&emsp;&emsp;&emsp;$X_i$ = Random Variable<br>
&emsp;&emsp;&emsp;&emsp;$\overline X$ = Mean of the distribution <br>
&emsp;&emsp;&emsp;&emsp;$\sigma$ = Standard Deviation of the distribution<br>
</p>
<br>

$$
Pearson's Median Skewness = 3 x \frac{Mean - Median}{Standard Deviation}
$$

<h1> Quantile-Quantile Plot (QQ Plot)</h1>
<hr>
<h3>Definition:</h3>

<p>The quantile-quantile (q-q) plot is a graphical technique for determining if two data sets come from populations with a common distribution.
A q-q plot is a plot of the quantiles of the first data set against the quantiles of the second data set. By a quantile, we mean the fraction (or percent) of points below the given value. That is, the 0.3 (or 30%) quantile is the point at which 30% percent of the data fall below and 70% fall above that value.</p>

<p>A 45-degree reference line is also plotted. If the two sets come from a population with the same distribution, the points should fall approximately along this reference line. The greater the departure from this reference line, the greater the evidence for the conclusion that the two data sets have come from populations with different distributions.</p>

<p>The q-q plot is formed by:</p>

 - Vertical axis: Estimated quantiles from data set 1
 - Horizontal axis: Estimated quantiles from data set 2
 
<p>The q-q plot is used to answer the following questions:</p>

 - Do two data sets come from populations with a common distribution?
 - Do two data sets have common location and scale?
 - Do two data sets have similar distributional shapes?
 - Do two data sets have similar tail behavior?

```python
import scipy.stats as stat
import pylab
def plot_data(df,feature):
    plt.figure(figsize = (10,3))
    plt.subplot(1,2,1)
    df[feature].hist()
    plt.xlabel(feature)
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist='norm',plot=pylab)
    plt.xlabel(feature)
    plt.show()
```

```python
for i in data.select_dtypes(include=['int64']).columns:
    plot_data(data,i)
```

<h1> Box Cox </h1>
<hr>
<h3>Definition:</h3>

<p>A Box Cox transformation is a transformation of <i>non-normal dependent variables into a normal shape</i>. Normality is an important assumption for many statistical techniques; if your data isnt normal, applying a Box-Cox means that you are able to run a broader number of tests.</p>
<p>The Box-Cox transformation transforms our data so that it closely resembles a normal distribution.
In many statistical techniques, we assume that the errors are normally distributed. This assumption allows us to construct confidence intervals and conduct hypothesis tests. By transforming your target variable, we can (hopefully) normalize our errors (if they are not already normal).
Additionally, transforming our variables can improve the predictive power of our models because transformations can cut away white noise.
</p>

<br><img src =https://www.researchgate.net/publication/324686305/figure/fig4/AS:864038716207104@1583014291597/Data-distributions-before-and-after-Box-Cox-transformation.jpg width = 700px height = 700px><br>

<p>Box-cox Transformation only cares about computing the value of $\lambda$  which varies from  5 to 5. A value of $\lambda$  is said to be best if it is able to approximate the non-normal curve to a normal curve.</p>
<p>If w is our transformed variable and y is our target variable, then</p>
<br>
$$
\omega_t = \left\{
    \begin{array}\\
        log(y_t) & \mbox{if } \lambda = 0;\\
        (y^\lambda_t -1)/\lambda & \mbox{otherwise.}
    \end{array}
    \right.
$$
<br>
<p>
where, <br>
&emsp;&emsp;&emsp;&emsp;t= time period<br>
&emsp;&emsp;&emsp;&emsp;$\lambda$ = parameter that we choose<br>
</p>
<br>
<p><b>NOTE:</b> We choose the value of lambda that provides the best approximation for the normal distribution of our response variable.</p>
    
<p>Scipy has a boxcox function that will choose the optimal value of lambda for us.</p><br>
$$
scipy.stats.boxcox()
$$
<br>
<p>
Simply pass a 1-D array into the function and it will return the Box-Cox transformed array and the optimal value for lambda. You can also specify a number, alpha, which calculates the confidence interval for that value. (For example, alpha = 0.05 gives the 95% confidence interval).
</p>
<br>
<p>Next, fit your model to the Box-Cox transformed data. However, you must revert your data to its original scale when you are ready to make predictions.<br>
For example, your model might predict that the Box-Cox transformed value, given other features, is 1.5. You need to take that 1.5 and revert it to its original scale (the scale of your target variable).</p><br>
$$
scipy.special.inv\_boxcox(y, lambda)
$$
<br>
<p>Pass in the data to be transformed, y, and the lambda with which you had transformed your data.</p>
